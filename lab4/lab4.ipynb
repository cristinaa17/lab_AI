Ex 1 - Import dataset

```python
import pandas as pd

dataset = pd.read_csv("data.csv")
(dataset.head())
```

Ex 2 - Jucători cu vârsta > 40

```python
players_over_40 = dataset[dataset["Age"] > 40]
players_over_40.head(10)
```

Ex 3 - Jucători cu Overall ≥ 85 și Age < 25

```python
top_players = dataset[(dataset["Overall"] >= 85) & (dataset["Age"] < 25)]
top_players
```

Ex 4 - Sortare după Skill Moves descrescător

```python
sorted_players = dataset.sort_values(by="Skill Moves", ascending=False)
sorted_players
```

Ex 5: Jucători cu contract până în 2021

```python
dataset["Contract Valid Until"] = dataset["Contract Valid Until"].astype(str)
dataset[dataset["Contract Valid Until"].str.contains("2021", na=False)]
```

Ex 6 - Dimensiunea dataset-ului și jucători unici

```python
rows, cols = dataset.shape
print("Numar randuri:", rows)
print("Numar coloane:", cols)

unique_players = dataset["Name"].nunique()
print("Numar jucatori unici:", unique_players)
```

Ex 7 - Top naționalități

```python
most_common = dataset["Nationality"].value_counts().idxmax()
print("Cea mai frecventa nationalitate:", most_common)

top5 = dataset["Nationality"].value_counts().head(5)
print("Top 5 nationalitati:")
print(top5)
```

Ex 8 - Pie chart naționalități (Top 5)

```python
import matplotlib.pyplot as plt

top5 = dataset["Nationality"].value_counts().head(5)

top5.plot.pie(autopct='%1.1f%%')
plt.title("Top 5 naționalități")
plt.ylabel("")
plt.show()
```

Ex 9 - Media atributelor pe naționalitate

```python
dataset.groupby("Nationality")[["SprintSpeed", "Acceleration"]].mean()
```

Ex 10 - Completare valori lipsă Position

```python
dataset["Position"] = dataset["Position"].fillna("Unknown")
```

Ex 11 - Club cu cea mai mare medie Overall

```python
club_mean = dataset.groupby("Club")["Overall"].mean()

best_club = club_mean.idxmax()
best_value = club_mean.max()

print("Club:", best_club)
print("Media Overall:", best_value)
```

Ex 12 - Jucători cu Value > Wage

```python
def convert_money(x):
    if isinstance(x, str):
        x = x.replace('€', '').replace('K', '000').replace('M', '000000')
        return float(x)
    return x

dataset["Value"] = dataset["Value"].apply(convert_money)
dataset["Wage"] = dataset["Wage"].apply(convert_money)

count = (dataset["Value"] > dataset["Wage"]).sum()

print("Numar jucatori:", count)
```

Ex 13 - Coloană is_underpaid

```python
dataset["is_underpaid"] = dataset["Wage"] < (dataset["Value"] / 100)

dataset[["Name", "Wage", "Value", "is_underpaid"]].head()
```

Ex 14 - Scor general jucători

```python
dataset["Score"] = (
    0.3 * dataset["Overall"] +
    0.3 * dataset["Potential"] +
    0.2 * dataset["SprintSpeed"] +
    0.2 * dataset["ShortPassing"]
)

dataset[["Name", "Score"]].head()
```

Ex 15 - Jucători "afacere bună"

```python
good_deals = dataset[["Name", "Wage", "Value"]].copy()

good_deals["difference"] = good_deals["Value"] - good_deals["Wage"]

good_deals = good_deals.sort_values(by="difference", ascending=False)

good_deals
```

Ex 16 - Scatterplot Wage vs Value

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=dataset, x="Wage", y="Value")

plt.title("Wage vs Value")
plt.show()
```

Ex 17 - BONUS: Hover pe grafic

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool

source = ColumnDataSource(dataset)

p = figure(title="Wage vs Value (Hover enabled)")

p.scatter("Wage", "Value", size=8, source=source)

hover = HoverTool(tooltips=[
    ("Name", "@Name"),
    ("Wage", "@Wage"),
    ("Value", "@Value")
])

p.add_tools(hover)

show(p)
```
